```python
def is_parfait(n, p):
    if (pow(2,p-1)*(pow(2,p)-1))==n:
        return True
    else: 
        return False
```

```python
def dicho_iteratif(x, n):
    y = 1
    while (n > 0):
        if (n%2 == 0):
            x = x*x
            n = n/2
        else: 
            y = y*x
            x = x*x
            n = (n-1)/2
    return y
```
